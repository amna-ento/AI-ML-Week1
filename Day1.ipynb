```python
df["TotalCharges"] = pd.to_numeric(
    df["TotalCharges"],
    errors="coerce"
)
```

```python
import matplotlib.pyplot as plt

missing = df["TotalCharges"].isnull().sum()
not_missing = df["TotalCharges"].notnull().sum()

plt.bar(["Missing", "Not Missing"], [missing, not_missing])
plt.title("Missing Values in TotalCharges")
plt.ylabel("Count")
plt.show()
```
